```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('pagoda.jpg')
bmg = data [70:330, 70:550]
print(data.shape)
zoom = nd.zoom(bmg, (5,5,1))
print(zoom.shape)
plt.imshow(bmg)
plt.show()
plt.imsave('ngoichuazoom5.jpg', zoom)
plt.imshow(zoom)
plt.show()

```
