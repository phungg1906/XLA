```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('quang_ninh.jpg')
bmg = data [50:330, 400:700]
print(data.shape)
d1 = nd.rotate(bmg, 45)

data = iio.imread('quang_ninh.jpg')
bmg1 = data [330:400, 550:700]
d2 = nd.rotate(bmg1, 45, reshape=False)

plt.imsave('nuixoay45.jpg', d1)
plt.imshow(d1)
plt.show()
plt.imsave('thuyenxoay45.jpg', d2)
plt.imshow(d2)
plt.show()
```
