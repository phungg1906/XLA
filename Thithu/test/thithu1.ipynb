```python
import cv2
import numpy as np
import random
import matplotlib.pylab as plt

fruits = cv2.imread('colorful-ripe-tropical-fruits.jpg')
h, w = fruits.shape[:2] # (w, h, 3) lay w và h
resized_fruits = cv2.resize(fruits, (w + 30, h + 30)) #+30 moi chieu
cv2.imwrite('output_fruits_resized.jpg', resized_fruits)

quang_ninh = cv2.imread('quang_ninh.jpg')
h2, w2 = quang_ninh.shape[:2]
center = (w2 // 2, h2 // 2) # lấy trung điểm
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(quang_ninh, rotation_matrix, (w2, h2))
flipped = cv2.flip(rotated, 1)  
cv2.imwrite('output_quangninh_rotated_flipped.jpg', flipped)


pagoda = cv2.imread('pagoda.jpg')
enlarged = cv2.resize(pagoda, (0, 0), fx=5.0, fy=5.0)
blurred = cv2.GaussianBlur(enlarged, (7, 7), 0)
cv2.imwrite('output_pagoda_enlarged_blurred.jpg', blurred)

plt.figure(figsize=(15, 10))

# Ảnh 1 - Fruits resized
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(resized_fruits, cv2.COLOR_BGR2RGB))
plt.title("Fruits resized (+30px)")
plt.axis('off')

# Ảnh 2 - Quang Ninh xoay + lật

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.title("Quang Ninh rotated + flipped")
plt.axis('off')

# Ảnh 3 - Pagoda phóng to
plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title("Pagoda enlarged x5")
plt.axis('off')
```
