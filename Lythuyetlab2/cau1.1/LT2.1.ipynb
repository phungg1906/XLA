```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# M nh v chuyn sang grayscale
img = Image.open('bird.png').convert('L')

# Chuyn nh thnh mng numpy
im_1 = np.asarray(img)

# o nh (inversion): gi tr pixel mi = 255 - gi tr pixel c
im_2 = 255 - im_1

# To nh mi t mng numpy
new_img = Image.fromarray(im_2)

# Hin th nh gc
img.show()

# Hin th nh o ngc dng matplotlib
plt.imshow(new_img, cmap='gray')
plt.title("Inverted Image")
plt.axis('off')
plt.show()
```
