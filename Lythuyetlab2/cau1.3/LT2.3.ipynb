```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math

# M nh v chuyn sang grayscale
img = Image.open('pagoda.jpg').convert('L')

# Chuyn nh thnh mng numpy
im_1 = np.asarray(img)

# Chuyn kiu d liu t int sang float
b1 = im_1.astype(float)

# Tm gi tr ln nht trong nh
b2 = np.max(b1)

# p dng php bin i logarit
# c = 128.0 * log(1 + pixel) / log(1 + max_pixel)
c = (128.0 * np.log(1 + b1)) / np.log(1 + b2)

# Chuyn v kiu nguyn int  hin th
c1 = c.astype(np.uint8)

# To nh mi t mng  bin i
d = Image.fromarray(c1)

# Hin th nh gc
img.show()

# Hin th nh  log-transform
d.show()

# Hoc dng matplotlib  hin th
plt.imshow(d, cmap='gray')
plt.title("Log Transformed Image")
plt.axis('off')
plt.show()
```
