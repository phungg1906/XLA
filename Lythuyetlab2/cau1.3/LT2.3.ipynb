```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math

# Mở ảnh và chuyển sang grayscale
img = Image.open('pagoda.jpg').convert('L')

# Chuyển ảnh thành mảng numpy
im_1 = np.asarray(img)

# Chuyển kiểu dữ liệu từ int sang float
b1 = im_1.astype(float)

# Tìm giá trị lớn nhất trong ảnh
b2 = np.max(b1)

# Áp dụng phép biến đổi logarit
# c = 128.0 * log(1 + pixel) / log(1 + max_pixel)
c = (128.0 * np.log(1 + b1)) / np.log(1 + b2)

# Chuyển về kiểu nguyên int để hiển thị
c1 = c.astype(np.uint8)

# Tạo ảnh mới từ mảng đã biến đổi
d = Image.fromarray(c1)

# Hiển thị ảnh gốc
img.show()

# Hiển thị ảnh đã log-transform
d.show()

# Hoặc dùng matplotlib để hiển thị
plt.imshow(d, cmap='gray')
plt.title("Log Transformed Image")
plt.axis('off')
plt.show()
```
